```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Bangalore_OLA_1_Lakh_Rides.csv')
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
```

```python
df.head(50)
```

```python
df.loc[df['booking_status'] != 'Success' , ['avg_vtat','avg_ctat']] = 0 
```

```python
df.loc[df['booking_status']!='Success' , ['driver_rating','customer_rating']] = 0
```

```python
df.loc[df['booking_status']=='Success' , ['incomplete_ride_reason']] = 'not_applicable'
df.loc[df['booking_status']=='Success' , ['reason_for_cancel_by_customer','reason_for_cancel_by_driver']] = 'not_applicable'


cancel_reason_columns =['reason_for_cancel_by_customer','reason_for_cancel_by_driver','incomplete_ride_reason'] 

for col in cancel_reason_columns:
    df[col] = df[col].fillna('not_known')
```

```python
value_and_rating_cols=['ride_distance']
for col in value_and_rating_cols:
    df[col]=df[col].fillna(0)
```

```python
df['booking_value'] = df['booking_value'].fillna(df['booking_value'].mean())
```

```python
df['customer_rating'].value_counts()
```

```python
df[['avg_vtat','avg_ctat','booking_value','ride_distance','driver_rating','customer_rating']] = df[['avg_vtat','avg_ctat','booking_value','ride_distance','driver_rating','customer_rating']].round(1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
df['year'] = df['date'].dt.year
df['day'] = df['date'].dt.day
df['month_name'] = df['date'].dt.month_name()
```

```python
df['year'].value_counts()
```

```python
df['vehicle_type'].unique()

print('-'*50)

df['vehicle_type'].value_counts()
```

```python
df['booking_status_intType'] = (df['booking_status'] == 'Success').astype(int)
```

```python
df['booking_status_intType'].mean()*100
```
